```python
# Check for outliers using the IQR method
def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outlier_mask = (data < lower_bound) | (data > upper_bound)
    return outlier_mask
```

An R-squared value of 1 means that the independent variables explain all the variability of the dependent variable around its mean.
In this Case, an R-squared value of 1 indicates that the model perfectly fits the data. This means that the model explains 100% of the variance in the 'Price' based on the features in X. In other words, the predictions made by the model are exactly matching the actual 'Price' values in the test set.

However, a perfect R-squared value (especially in practical scenarios) might also be a sign of overfitting, where the model has learned the training data too well, including its noise and outliers, and might not generalize well to unseen data.
